```python
import pandas as pd
import matplotlib as plt
import numpy as np
```

Data Reading

```python
df1 = pd.read_csv('Global_Education.csv', encoding=('latin-1'))
df1.head()
```

```python
df2 = pd.read_csv('ODEF_v2_1.csv')
df2.head()
```

Merging

```python
df1['ID'] = range(len(df1))
df2['ID'] = range(len(df2))

enriched_df = pd.merge(df1, df2, on='ID', how='inner')
enriched_df.head()
```

Eploring Merged Data

```python
print(enriched_df.head())
```

```python
print(enriched_df.info())
```

```python
print(enriched_df.describe())
```

Data Cleaning

```python
print(enriched_df.columns)
```

Confirming Non_numeric columns

```python
non_numeric_columns = enriched_df.select_dtypes(exclude=['number']).columns
print(non_numeric_columns)
```

Cleaning Non_numeric columns

```python
enriched_df['Countries and areas'] = enriched_df['Countries and areas'].str.strip()
enriched_df.head()
```

```python
enriched_df['Facility_Type'].fillna('Unknown', inplace=True)
enriched_df.head()
```

```python
enriched_df = pd.get_dummies(enriched_df, columns=['Facility_Name'])
enriched_df
```

Droping unnecessary columns

```python
enriched_df_cleaned = enriched_df.drop(['Source_ID', 'Geo_Source'], axis=1)
enriched_df.head()
```

Handling Duplicates

```python
enriched_df_cleaned = enriched_df_cleaned.drop_duplicates()
enriched_df.head()
```

Exploring Data Distribution

```python
import matplotlib.pyplot as plt
OOSR_Pre0Primary_Age_Male = enriched_df['OOSR_Pre0Primary_Age_Male']
mean_value = OOSR_Pre0Primary_Age_Male.mean()
fig, ax = plt.subplots(figsize=(7, 5))
ax.hist(OOSR_Pre0Primary_Age_Male, bins='auto')
ax.axvline(mean_value, color='black')
plt.ylabel('No. of pupils')
plt.xlabel('OOSR Pre0Primary Age Male')
plt.title('OOSR Pre0Primary Age Male')
plt.show()
```

Correlation

```python
numeric_columns = enriched_df.select_dtypes(include='number')
correlation_matrix = numeric_columns.corr()
max_corr_column = correlation_matrix['OOSR_Pre0Primary_Age_Male'].sort_values(ascending=False).index[1]
max_corr_value = correlation_matrix.loc['OOSR_Pre0Primary_Age_Male', max_corr_column]

print("Most Positively Correlated Column:", max_corr_column)
print("Max Correlation Value:", max_corr_value)
```

```python
numeric_columns = enriched_df.select_dtypes(include='number')
correlation_matrix = numeric_columns.corr()

min_corr_column = correlation_matrix['OOSR_Pre0Primary_Age_Female'].sort_values(ascending=True).index[0]
min_corr_value = correlation_matrix.loc['OOSR_Pre0Primary_Age_Female', min_corr_column]

print("Most Negative Correlated Column:", min_corr_column)
print("Min Correlation Value:", min_corr_value)
```
